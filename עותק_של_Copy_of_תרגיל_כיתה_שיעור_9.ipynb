````python
# Global variable
total_count = 0

def add_item(count):
    global total_count
    total_count += count
    print(f"Added {count} items. Total now: {total_count}")

def remove_item(count):
    global total_count
    total_count -= count
    print(f"Removed {count} items. Total now: {total_count}")

def reset_count():
    global total_count
    total_count = 0
    print("Count reset to 0")

# Simulating various operations
add_item(5)
remove_item(2)
reset_count()
add_item(3)
remove_item(1)
````

````python
#   
def calculate_median(data):
    sorted_data = sorted(data)
    length = len(sorted_data)
    if length % 2 != 0:
        median = sorted_data[length // 2]
    else:
        mid1 = sorted_data[(length - 1) // 2]
        mid2 = sorted_data[(length + 1) // 2]
        median = (mid1 + mid2) / 2
    return median
````

````python
#   
def commons_table(data):
    salary_counts = {}

    for salary in data:
        if salary in salary_counts:
            salary_counts[salary] += 1
        else:
            salary_counts[salary] = 1

    for salary, count in salary_counts.items():
        print("salary: {}, count: {}".format(salary, count)) #         

    return salary_counts
````
