````python
#    
def category(num):
    if num < 0.34:
        return 'small'
    elif num < 0.51:
        return 'mid'
    else:
        return 'big'

df['size'] = df['bone_length'].apply(category)
````

````python
#     
x = df['size'].value_counts()
````

````python
#       
sns.set(style="whitegrid")
sns.countplot(x="size", data=df, palette="Set3")
````
