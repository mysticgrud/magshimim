```python
# @title טעינת ספריות
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# @title המחשה חלוקה למחלקות- הריצו - שחקו עם רוחב המחלקה וענו על השאלה

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets


data = np.array([np.round(np.random.normal(10, 2), 2) for _ in range(1000)])

def plot_hist(data, bins, ax):
    sns.histplot(data, bins=bins, color="#64edbfff", ax=ax, stat="density", legend=False)
    ax.set_title(f'Number of Bins: {bins}')
    ax.set(ylabel='')

@interact(bins=widgets.IntSlider(min=1, max=1000, step=1, value=5))
def interactive_histogram(bins):
    fig, ax = plt.subplots(figsize=(8, 6))
    plot_hist(data=data, bins=bins, ax=ax)
    plt.show()
```

חקרו את הצורות השונות של התפלגות הנתונים בהדגמה הבאה , הסבירו מה מיוחד בצורה ומה זה אומר על הנתונים במקרה זה? מה זה אומר על שכיחות הערכים? איך הם מתפלגים?
האם זה נותן מידע גם על מדדים כלשהם לפי צורה?

```python
# @title הריצו לקבל הצורות השונות של ההיסטוגרמה

import random
right_tail = np.array([np.round(random.betavariate(2,10),2) for i in range(1000)])
left_tail = np.array([np.round(random.betavariate(10,2),2) for i in range(1000)])
uniform = np.array([np.round(random.uniform(0,10), 2) for i in range(15000)])
normal = np.array([np.round(random.gauss(10,2),2) for i in range(1000)])

right = np.array([np.round(random.betavariate(2,10),2) for i in range(1500)])
left = np.array([np.round(random.betavariate(10,2),2) for i in range(1500)])
multimodal = np.concatenate((right, left), axis = None)

def plot_hist(data, title=None, range=False, iqr=False, std=False, hist_color="#f1f3feff", kde_color="#141c3aff", ax=None, bins=None):
    if not ax:
        ax = plt.axes()
    if title:
        ax.set_title(title[::-1])  # reverse hebrew titles
    if bins:
        sns.histplot(data, color=hist_color, ax=ax, stat="density", legend=False, bins=bins)
    else:
        sns.histplot(data, color=hist_color, ax=ax, stat="density", legend=False)
    sns.kdeplot(data, color=kde_color, lw=5, alpha=0.5, ax=ax)
    if iqr:
        q1 = np.quantile(data, 0.25)
        q2 = np.quantile(data, 0.5)
        q3 = np.quantile(data, 0.75)
        ax.axvline(x=q1, c='#64edbfff', lw=4, ls="-")
        ax.axvline(x=q2, c='#ecc505ff', lw=4, ls="-")
        ax.axvline(x=q3, c='#e94f4b', lw=4, ls="-")
    if range:
        max_limit = max(data)
        min_limit = min(data)
        ax.axvline(x=max_limit, c='#64edbfff', lw=4, ls="-")
        ax.axvline(x=min_limit, c='#64edbfff', lw=4, ls="-")
    if std:
        avg_value = np.average(data)
        std_dev = np.std(data)
        ax.axvline(avg_value, 0, 1, c='#e94f4bff', lw=4, ls="-")
        avg_std_up = avg_value + std_dev
        avg_std_down = avg_value - std_dev
        ax.axvline(avg_std_up, 0, 1, c='#6400e4', lw=1, ls="--")
        ax.axvline(avg_std_down, 0, 1, c='#6400e4', lw=1, ls="--")
    ax.set(ylabel='')

def interactive_histogram(hist_type='Right Tail', bins=30, range=False, iqr=False, std=False):
    fig, ax = plt.subplots(figsize=(8, 6))

    if hist_type == 'Right Tail':
        data = right_tail
    elif hist_type == 'Normal':
        data = normal
    elif hist_type == 'Left Tail':
        data = left_tail
    elif hist_type == 'Uniform':
        data = uniform
    elif hist_type == 'Multimodal':
        data = multimodal
    else:
        raise ValueError('Invalid histogram type')

    plot_hist(data=data, bins=bins, range=range, iqr=iqr, std=std, ax=ax)
    plt.show()

# Creating interactive widgets
hist_type_dropdown = widgets.Dropdown(
    options=['Right Tail', 'Normal', 'Left Tail', 'Uniform', 'Multimodal'],
    value='Right Tail',
    description='Histogram Type:'
)

bins_slider = widgets.IntSlider(min=1, max=100, step=1, value=30, description='Bins:')
range_checkbox = widgets.Checkbox(value=False, description='Show Range')
#iqr_checkbox = widgets.Checkbox(value=False, description='Show IQR')
#std_checkbox = widgets.Checkbox(value=False, description='Show Std Dev')

# Defining the interactive function
@interact(hist_type=hist_type_dropdown, bins=bins_slider, range=range_checkbox)#, iqr=iqr_checkbox, std=std_checkbox)
def interactive_plot(hist_type, bins, range): #, iqr, std):
    interactive_histogram(hist_type, bins, range)#, iqr, std)
```

<br>

זנב ימין(Right Tail) - יש שכחיות גבוהה יותר של ערכים נמוכים הם מתפלגים לכיוון שמאל וזה אומר על המדדים שהחציון כניראה יהיה נמוך יותר מהממוצע

זנב שמאל(left Tail) - יש שכיחות גבוהה יותר של ערכים גבוהיים הם מתפלגים לכיוון ימין וזה אומר על המדדים שהחציון כניראה יהיה גבוהה יותר מהממוצע

התפלגות נורמלית (Normal)- התפלגות נורמאלית כל ערכי האמצע שווים ויש שכיחות גבוהה של ערכים אמצעיים, לא גבוהים מידי ולא נמוכים מידי

מלבנית (Uniform)- יש שכיחות זהה להכל

גמל (Multimodal)- זה כאשר יש שני "הרים" בהתפלגות של הנתונים או יותר
