מצורפת רשימת הקניות השבועית של מספר תלמידי מגשימים. בנוסף מצורפת רשימה של כל הפירות הקיימים בסופר.
<br>
מצורפת גם רשימת הקניות של אדם, חניך במגשימים, עליכם לבדוק כמה פריטים מתוך רשימת הקניות של אדם הם פירות.
<br><br>
<b>הוראות:</b>
<br>
1. צרו רשימה חדשה של הפירות אשר כל האותיות בה הן קטנות (חשבו, למה בעצם צריך ליצור רשימה כזו?)
<br><br>
2. בדקו עבור כל פריט ברשימה של אדם האם הוא מופיע ברשימת הפירות (דוגמא להאם פריט מופיע ברשימה מופיעה למטה)
<br><br>
3. הדפיסו את מספר הפירות שיש לתלמיד בסל

````python
#בקטע הקוד הבא נוכל לראות האם פריט מסויים קיים ברשימה או לא
lst=[ 1, 6, 3, 5, 3, 4 ]
num = 5
if num in lst:
    print("exist")
else:
    print("not exist")

# הדוגמא הזו היא מספרית, אך הדבר יעבוד גם עבור מחרוזת. שימו לב כי הערכי המחזורות צריכים להיות זהים בידיוק
````

````python
fruits = ["Apple", "Banana", "Grapes", "Orange", "Strawberries", "Watermelon", "Kiwi",
          "Pineapple", "Peaches", "Blueberries", "Mango", "Papaya"]

adam = ["apple", "mango", "banana", "kiwi", "bread"]

## השלימו את קטע הקוד כאן
fruits_S = ["apple", "banana", "grapes", "orange", "strawberries", "watermelon", "kiwi", "pineapple", "peaches", "blueberries", "mango", "papaya"]

counter = 0
while counter < 5:
    if adam[counter] in fruits_S:
        print("exist")
    else:
        print("not exist")
    counter += 1


````
