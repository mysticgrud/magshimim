רמז:

1) תמצאו את הממוצע של כל אחד מהנתונים הנ"ל

````python
#השלימו כאן קטע קוד
#------------
mean_hp = df['HP'].mean()
mean_Defense = df['Defense'].mean()
mean_sp_def = df['Sp. Def'].mean()
high = df[((df['HP']> mean_hp) & (df['Defense']> mean_Defense)) & (df['Sp. Def']> mean_sp_def)]
high
#------------
````

<br>
אה, עוד דבר- יש בעולם הפוקימונים מחסור חמור בזכרון, אז מכיוון שיצרנו כבר מקודם DataFrame חדש מסונן, השתמשו בפרמטר שלמדנו כדי לבצע את המחיקות שנצטרך לבצע עכשיו על הDataFrame הקיים שיצרנו מקודם, ושלא יווצר DataFrame חדש.

````python
#השלימו כאן קטע קוד
#------------
high = high[(high['Type 1'] ==  'Water') | (high['Type 2']== 'Water')]
high
#------------
````

<div dir=rtl>
<font size='4'>
נרצה עכשיו לראות מה עדיף- שהסוג הראשי יהיה מים או הסוג המשני יהיה מים.
<br>
צרו עמודה חדשה הנקראת Defensive_Total  שערכיה הם הסכום הבא: HP + Defense + Sp. Def
<br>
לאחר מכן,
קבצו בבקשה את הפוקימונים לפי הסוג ראשי וחשבו את החציון של Defensive_Total לכל קבוצה והוציאו מהנתונים שקיבלתם את החציון של הקבוצה שהסוג הראשי שלה הוא מים.
<br>
עשו את אותו הדבר גם לפי הסוג המשני.

````python
#השלימו כאן קטע קוד
#------------
high['Defensive_Total'] = high['HP'] + high['Defense'] + high['Sp. Def']
median1 = high[high['Type 1'] == 'Water']['Defensive_Total'].median()
median2 = high[high['Type 2'] == 'Water']['Defensive_Total'].median()

print("Median Defensive Total for Type 1 'Water':", median1)
print("Median Defensive Total for Type 2 'Water':", median2)
#------------
````

<div dir=rtl>
<font size='4'>
צרו DataFrame חדש עם עמודת median המכילה את 2 הערכים שחושבו מקודם (ושהלייבלים יהיו בהתאם) וצרו גרף עמודות המציג אותו.
<br>
כלומר צרו גרף עמודות לכל סוג של TYPE לפי החציון שחישבת קודם.( העזרו בDF חדש לכל עמודה של חציון עם תגית מואמת)

````python
#השלימו כאן קטע קוד
#------------
ex2 = high.copy().reset_index(drop=True)
ex2['median'] = None
ex2.loc[ex2['Type 1'] == 'Water', 'median'] = median1
ex2.loc[ex2['Type 2'] == 'Water', 'median'] = median2
sns.barplot(data = ex2, x = ex2['Type 1'])
#------------
````
