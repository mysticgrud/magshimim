````python
# Global variable
total_count = 0

def add_item(count):
    global total_count
    total_count += count
    print(f"Added {count} items. Total now: {total_count}")

def remove_item(count):
    global total_count
    total_count -= count
    print(f"Removed {count} items. Total now: {total_count}")

def reset_count():
    global total_count
    total_count = 0
    print("Count reset to 0")

# Simulating various operations
add_item(5)
remove_item(2)
reset_count()
add_item(3)
remove_item(1)
````

````python
salary_df = pd.DataFrame(data, columns=["Age" , "Gender","Education Level", "Job Title", "Years of Experience", "Salary" ]) #  יש להשלים את עמודות של הטבלה על פי הסדר בקובץ הנתונים שעבדתך איתו קודם!

# calculate_average_salary יש להשלים כאן את הפונקציה לחישוב הממוצע
def calculate_average_salary(data):
  cnt = 0
  total = 0
  total_salary = 0
  for row in data:
      total_salary += row[-1]
      average_salary = total_salary / len(data)
  return average_salary
print(calculate_average_salary(data))
````

````python
#@title הקוד המוסתר כאן, לחץ על חץ ההרצה פשוט

#הריצו את קטע הקוד הבא
# Creating dropdowns for the interface
gender_dropdown = widgets.Dropdown(
    options=['All'] + list(salary_df['Gender'].unique()),
    description="Gender:"
)

education_dropdown = widgets.Dropdown(
    options=['All'] + list(salary_df['Education Level'].unique()),
    description="Education Level:"
)

output_text = widgets.Output()

````

````python
def calculate_and_display_salary(change):
    with output_text:
        clear_output(wait=True)
        selected_gender = gender_dropdown.value
        selected_education = education_dropdown.value

        # Filtering the data
        filtered_data = salary_df
        if selected_gender != 'All':
            filtered_data = filtered_data[filtered_data["Gender"] == selected_gender]  # השלימו את העמודות בהתאם לתנאי
        if selected_education != 'All':
            filtered_data = filtered_data[filtered_data["Education Level"] == selected_education] # השלימו את העמודות בהתאם לתנאי
        filtered_data = filtered_data.to_numpy().tolist()

        # Using your function to calculate average salary
        average_salary = calculate_average_salary(filtered_data)
        print(f"The average salary for {selected_gender} with {selected_education} education is {average_salary}")

# Linking dropdowns to the function
gender_dropdown.observe(calculate_and_display_salary, names='value')
education_dropdown.observe(calculate_and_display_salary, names='value')

# Display widgets
display(gender_dropdown, education_dropdown, output_text)
````

````python
#השלימו כאן שימוש בפונקציה
from collections import Counter

def calculate_common_salary(data):
  salaries = [row[-1] for row in data]
  salaries_counter = Counter(salaries)
  most_common = salaries_counter.most_common(1)[0][0] #most common in postion 1 thats y the (1) and [0][0] is supposed the extract that from the salaries thing according to google
  return most_common
````

````python
#השלימו כאן שימוש בפונקציה
def calculate_median(data):
    sorted_data = sorted(data)
    length = len(sorted_data)
    if length % 2 != 0:
        median = sorted_data[length // 2]
    else:
        mid1 = sorted_data[(length - 1) // 2]
        mid2 = sorted_data[(length + 1) // 2]
        median = (mid1 + mid2) / 2
    return median
````
