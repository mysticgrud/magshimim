לפניכם רשימה מקוננת של מספרים (מטריצה), עליכם לסכום כל עמודה של המטריצה ולהדפיס את התוצאה. לדוגמא סכום העמודה הראשונה הוא 1+4+7 = 12 <br>
שימו לב שבקוד ישנם 2 טעויות, (סינטקס ולוגית), עליכם למצוא את הטעויות ולהריץ את הקוד התקין

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

mat_size = 3

for j in range(mat_size):
    col_sum = 0
    for i in range(mat_size):
        col_sum += matrix[i][j]
    print(f"Sum of Column {j + 1}: {col_sum}")
````

סעיף ו':</b></font><br><font size='4'>
לפניכם מידע על סטודנטים במכללת שק״ל (שקר כלשהו).<br>
 <b> חלק א': </b> עליכם לחשב את מממוצע הציונים של כל סטודנט. שימו לב ששלושתם הרכיבים הראשונים בעמודה הם שם גיל ומין ועליכם להשתמש בחיתוך עלל מנת לחשב את הממוצע. שימו לבה שעל מנת לחשב את הממוצע תוכלו להשתמש בפעולות sum ו len שלמדנו בשיעורים הקודמים.
<br><b> חלק ב': </b> עליכם לחשב את הממוצע הכיתתי ולהדפיס אותו.

````python
student_data = [
    ["Sara", 18, "F", 85, 90, 92],
    ["Alon", 20, "M", 78, 85, 80],
    ["Rotem", 19, "M", 92, 88, 95],
    ["David", 21, "M", 80, 75, 82],
    ["Lotem", 20, "F", 88, 90, 87]
]

## חלק א' חשבו את ממוצע הציונים לכל תלמיד
sumavg = 0
for student in student_data:
  sums = student[-1] + student[-2] + student[-3]
  stuavg = sums/3
  print(stuavg)
  sumavg += stuavg
## חלק ב' חשבו את הממוצע הכיתתי
avrg = sumavg/len(student_data)
print("the overall class average is: ", avrg)
````

````python
name_list = ["Alice", "Dana", "Charlie", "David", "Eva"]
## הקלידו כאן את קטע הקוד:
from collections import Counter
counter = Counter(''.join(name_list)) #מחבר את הכל לדבר אחד שאליו הוא יכול לעבור

print(counter['a']) #סופר כמה פעמים a מופיע
````
