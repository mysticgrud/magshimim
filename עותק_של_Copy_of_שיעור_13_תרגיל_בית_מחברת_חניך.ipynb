```python
dfg = pd.DataFrame({"name": ["sara", "rivka", "lea", "rachel"],
                   "Fav_type": ["kartiv", "artic", "artic", "kartiv"],
                   "Fav_taste": ["mint", "chocolate", "mint", "chocolate"]})

pd.pivot_table(data=dfg ,values="name" , index=["Fav_type"], columns=["Fav_taste"] , aggfunc=pd.Series.count)
```
