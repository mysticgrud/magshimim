פרמטר inplace
</b>

<div dir="rtl">
<font  size="4">
לפני שנמשיך, נדבר על הפרמטר inplace.
<br>
הפרמטר קיים במספר מתודות pandas ומשפיע על ההתנהגות של אופן הפעלת המתודה. הפרמטר מקבל ערך בוליאני- True או  False. (ברירת המחדל הוא False)
<br>
אם inplace= True המתודה תשנה את ה- DataFrame הקיים.
אם inplace= False  המתודה תיצור עותק חדש של ה-DataFrame עם השינויים.
<br>
נראה דוגמא לשימוש:
<br>
כאשר המתודה היא false כלומר יווצר עותק חדש


```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)

df.drop(0, inplace=False)

print(df)
```

<div dir="rtl">
<font  size="4">
כאשר המתודה היא true - כלומר יש שינוי בdf הקיים


```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)

df.drop(0, inplace=True)

print(df)
```

<b>
המתודה copy
</b>

<div dir="rtl">
<font  size="4">
כאשר אנו מסננים df ורוצים ליצור ממנה טבלה חדשה, או פשוט ליצור טבלה זהה לטבלת דאטה שלנו, מבלי ״לדרוס״ את הטבלה הנוכחית, נשתמש במתודה copy()
<br>
המתודה תחזיר עותק של DataFrame ע״י יצירה של "deep copy" בזכרון המחשב (מומלץ לקרוא על המונח
<a href="https://medium.com/@chandradip93/shallow-copy-and-deep-copy-are-two-different-approaches-to-copying-objects-or-data-structures-b00e9b3f45f5">כאן</a>
), המשמעות בפועל היא ששינוים על העותק שנוצר לא ישפיעו על העותק המקורי.
<br>
<br>
נראה דוגמא:

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}

df = pd.DataFrame(data)

#  ניצור עותק של טבלת הנתונים שלנו
df_copy = df.copy()

df_copy.query('Age > 25', inplace=True)

print("Original DataFrame:")
print(df)

print("\nModified DataFrame:")
print(df_copy)
```

<div dir=rtl>

כפי שלמדנו לכל DataFrame יש תכונה הנקראת index שהיא בעצם הlabel על כל השורות שלנו.
 מה שdf.index מחזיר לנו זה את התיאור של האינדקס של הDataFrame.
במקרה שלנו לDataFrame יש אינדקס מספרי המתחיל ב0 ונגמר ב923 וקופץ ב1 כל פעם. זה גם האינדקס הדיפולטי שDataFrame מקבל. בואו נראה מה קורה אם אנחנו נותנים לDataFrame אינדקס משלנו!

<div dir=rtl>
<font size =4>
בואו ניצור df חדש עם אינדקס מספרי.

```python
date_df = pd.DataFrame({'month': [1, 4, 7, 10],
                   'year': [2012, 2014, 2013, 2014],
                   'sale': [55, 40, 84, 31]})
date_df
```

<div dir=rtl>
<font size =4>
יצרנו DataFrame חדש של תאריכים, לא הגדרנו לו אינדקס לכן יהיה לו אינדקס מספרי הקופץ באחד לכל שורה. עכשיו בואו נשתמש בפעולה df.set_index ונראה מה היא עושה.

```python
month_date_df = date_df.set_index('month')
month_date_df
```

<div dir=rtl>
<font size =4>
הפעולה הזאתי שינתה את האינדקס שלנו להיות העמודה של החודש! אפשר להפעיל את הפעולה על כל עמודה שרוצים בdf ובכך לשנות את הDataFrame.

```python
month_date_df.index
```

<div dir=rtl>
<font size =4>
למעלה אנחנו רואים את התכונות של האינדקס החדש שלנו.

<div dir=rtl>
<font size =4>
עכשיו התחרטנו 😕 אנחנו כבר לא רוצים שהחודש יהיה בindex שלנו ואנחנו רוצים לחזור למספרים ההגיוניים שלנו. אל דאגה! pandas חשבו על זה והמציאו את הפעולה df.reset_index במיוחד בשביל זה. בואו נשתמש בה כדי להחזיר את המצב לקדמותו.

```python
regular_date_df= month_date_df.reset_index()
regular_date_df.index
```
