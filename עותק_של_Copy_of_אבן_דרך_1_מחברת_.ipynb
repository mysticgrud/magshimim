````python
##   
coat = [dog[8] for dog in data]
size = [dog[9] for dog in data]
####      

###      1
def coatL(coatt):
    if coatt == "short":
        return 1
    elif coatt == "medium":
        return 2
    elif coatt == "long" or coatt == "wirehaired":
        return 3

###      2
def sizeL(sizee):
    if sizee == "small":
       return 1
    elif sizee == "medium":
       return 2
    elif sizee == "large":
       return 3

## S
def perd_model(coatt, sizee):
  result_score = coatt + sizee ##   
  if result_score == 1 or result_score == 2:  ##   
     return "fast"
  elif result_score == 3 or result_score == 4:          ##   
    return "medium"
  else:
    return "slow"


##       
result = perd_model("short", "small")
print(result)
````

````python
# @title  
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Interface using IPywidgets
param1_dropdown = widgets.Dropdown(
    options=list(shelter_df[param1_name].unique()),
    description=param1_name+":"
)

param2_dropdown = widgets.Dropdown(
    options=list(shelter_df[param2_name].unique()),
    description= param2_name+":"
)

output_text = widgets.Output()
output_image = widgets.Image(value=open('default_image.jpg', 'rb').read(), format='jpg',layout=widgets.Layout(width='150px'))

def calculate_and_display_result(change):
    with output_text:
        clear_output(wait=True)

        selected_param1 = param1_dropdown.value
        selected_param2 = param2_dropdown.value

        #         
        #   XXX          
        score_param1 = coatL(selected_param1)
        score_param2 = sizeL(selected_param2)

        result_category = perd_model(score_param1, score_param2)
        print(f"Dog adoption time is {result_category}")

        # Display images based on the result
        image_path = f'{result_category}_adoption.jpg'
        output_image.value = open(image_path, 'rb').read()

# Link dropdowns to the function
param1_dropdown.observe(calculate_and_display_result, names='value')
param2_dropdown.observe(calculate_and_display_result, names='value')


# Display the interface
display(widgets.VBox([param1_dropdown, param2_dropdown, output_text, output_image]))

# Initial display
calculate_and_display_result(None)
````
