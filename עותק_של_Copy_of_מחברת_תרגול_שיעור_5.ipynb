מלבד האופרטורים שהכרנו עד עתה(+,-,*,/ וכן הלאה), ישנו אופרטור נוסף: %. לאופרטור הזה קוראים מודולו (modulo) והוא מאפשר חישוב של שארית החלוקה.<BR>
למשל, כאשר אנו מחלקים את המספר 10 ב-3, אנו מקבלים 3 ביחד עם <B> שארית 1</B>. השארית הזו היא בדיוק המודולו.<BR>
למידע והסבר נוספים תוכלו לחזור לסרטון שצפיתם בו בבית
<a href="https://www.youtube.com/watch?v=kzNhUy-m664&ab_channel=Lokipod">בקישור הבא.</a>
<BR>
כמה דוגמאות הרצה:

- החלוקה של 4 ב-2 לא משאירה שארית כלל (כי 4 מתחלק ב-2 באופן מלא) ולכן תוצאת המודולו תהיה 0:
</DIV>


````python
print('The modulo result of 4 % 2 is:', 4 % 2)
````

<div dir="rtl" align="right">
<font size='4'>

- החלוקה של 10 ב-3 תשאיר אותנו עם שארית 1 (כי שלוש "נכנס" בצורה שלמה 3 פעמים בתוך המספר 10, ונשאר עם שארית אחת), ולכן תוצאת המודולו תהיה 1:
</DIV>





````python
print('10 % 3 is: ', 10 % 3)
````

 אתם על סיפון חללית שעומדת לשגר לחלל למשימה מרגשת. המשימה שלכם היא להתחיל את ספירת ההשקה לאחור באמצעות לולאת WHILE כדי לספור לאחור מ-10 ל-1 לפני ההמראה.

משימה: כתבו תוכנית המשתמשת בלולאת WHILE כדי לספור לאחור מ-10 ל-1. הדפיסו כל מספר ספירה לאחור, וכשהספירה לאחור מגיעה ל-1, הדפיסו הודעה המציינת שהחללית מוכנה להמריא.

<font size='4'><br>
אתה מוצא את עצמך בבניין מסתורי עם מעלית קסומה שעוצרת רק בקומות עם מספרים זוגיים. כדי להגיע ליעד שלך, עליך לתכנת את המעלית באמצעות לולאת WHILE כדי להבטיח שאתה עוצר רק בקומות זוגיות.
<br>
משימה: כתוב תוכנית המדמה את תנועת המעלית באמצעות לולאת WHILE. התחילו מקומת הקרקע (קומה 0) והמשיכו לנוע לקומה הזוגית הבאה עד שתגיעו לקומה הרצויה  42.

````python
#השלימו את קטע הקוד כאן
floor = 0
while floor<=42:
  if floor%2==0:
    print(floor)
  floor+=1
````
