```python
#   
#------------
def weight_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    elif 30 <= bmi < 34.9:
        return "Obese"
    else:
        return "Extreamly Obese"
#------------
```

<div dir="rtl">

#          : 167 "  68 ".   

```python
#   
height = 167
weight = 68

category = weight_category(calculate_bmi(height, weight))

print(category)
```

<div dir="rtl">

#       LAMBDA            ( 167 "  68 ")

```python
#   

calculate_bmi_02 = lambda height, weight: weight / ((height/100) ** 2)

weight_category_02 = lambda bmi: "Underweight" if bmi < 18.5 else "Normal weight" if 18.5 <= bmi < 24.9 else "Overweight" if 25 <= bmi < 29.9 else "Obese" if 30 <= bmi < 34.9 else "Extremely Obese"

category = weight_category_02(calculate_bmi_02(height, weight))

print (category)
```

```python
#   
import pandas as pd

gender = pd.DataFrame(filtered_df['gender'].value_counts(normalize=True) * 100).reset_index()
gender.columns = ['gender', '%gender']
filtered_df = pd.merge(left=filtered_df, right=gender, how='inner', on=['gender'])

table = pd.pivot_table(filtered_df, index='weight_category', columns='%gender', values='gender', aggfunc='count')

table = table.fillna(0)
```
