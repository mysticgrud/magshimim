<font  size="4">
כתוב תוכנית Python לחישוב גיל הכלב בשנות הכלב.
הערה: בשנתיים הראשונות(של שנות אדם), שנת כלב שווה ל-10.5 שנות אדם. לאחר מכן, כל שנת כלב שווה ל-4 שנות אדם.
הגדר משתנה שקולט את גיל הכלב בשנות אדם וחשב והצג מהו גילו בשנות כלב. בדוק שהערכים שנקלטו תקינים


```
לדוגמה כלב בן 4 בשנות אדם , יהיה בן 29 שנות כלב לפי חישוב של 2*10.5 +2*4
```


</div>

````python
#השלם את הקוד שלך כאן עבור סעיף זה
human_years = int(input("enter ur dogs age in human years: "))
if human_years>2:
  dog_age = 2 * 10.5 + (human_years-2)*4
elif human_years == 2:
  dog_age = 2*10.5
else:
  print("not a valid dogs age, rerun the program and enter a valid one")
print("the dog age is: ", dog_age)
````

````python
#@title הקוד המוסתר כאן, לחץ על חץ ההרצה פשוט

import pandas as pd

def read_csv_from_googledrive(sheet_id,sheet_name):
  url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
  # Read the Google Sheets data into a Pandas DataFrame
  df = pd.read_csv(url)
  return df

def convert_df_to_list(df):
  data_list = []
  for value in df.values:
    data_list.append(list(value))
  return data_list


shelter_df = read_csv_from_googledrive("15YALKABON4VcLsVPBin7rLmy0OQ53gEeFV-Vyuk0dgs","ShelterDogs")
data = convert_df_to_list(shelter_df)
````

<div dir="rtl" align="right"><font color='#6400E4' size="6" ><b>5. מהו הגזע הכי נפוץ? </b></font><br>
<font  size="4" >
הקוד שלפניכם מייצר מילון שכיחויות, השלימו את הקטע הקוד:

````python
dogs_breed_dict = {}

for dog in data:
  dog_breed = dog[4]
  if dog_breed in dogs_breed_dict:
    ##השלם
    dogs_breed_dict[dog_breed]+=1
  else:
    ##השלם
    dogs_breed_dict[dog_breed]=1

print(dogs_breed_dict)
most_breed = max(dogs_breed_dict, key= dogs_breed_dict.get )
print("most common breed= ", most_breed)
````


<div dir="rtl" align="right"><font  size="4" >
בונוס: במחברת התרגול של שיעור 7 ראינו דרך נוספו ליצור מילון שכיחויות, נסו ליצור את המילון כעת באותו אופן שיצרנו בשיעור 7. רמז- כדאי להשתמש בפונקציית עזר מובנית

````python
from collections import Counter

#השלם כאן את הקוד
dogs_breed_counter = Counter()

for dog in data:
    dog_breed = dog[4]
    dogs_breed_counter[dog_breed] += 1

print(dict(dogs_breed_counter))
````

לאחר מכן נכתוב את פונקציות המודל שמקבלת את הניקוד שהתקבל עבור כל פרמטר. הפונקציה לפי טווח המתאים תחזיר את התשובה וקצב האימוץ. שימוש לב הפונקציה חייבת לקבל את <b>ניקוד</b> הפרמטרים שבחרתם ולהחזיר את קצה האימוץ בצורת <b>מחרוזת</b>

````python
## דוגמא ליצירת רשימה
coat = [dog[8] for dog in data]
size = [dog[9] for dog in data]
####  השלימו כאן את הפונקציות שלכם

### השלימו כאן את הפונקציה לפרמטר 1
def coatL(coatt):
    if coatt == "short":
        return 1
    elif coatt == "medium":
        return 2
    elif coatt == "long" or coatt == "wirehaired":
        return 3

### השלימו כאן את הפונקציה לפרמטר 2
def sizeL(sizee):
    if sizee == "small":
       return 1
    elif sizee == "medium":
       return 2
    elif sizee == "large":
       return 3

##פונקציית המודלS
def perd_model(coatt, sizee):
  result_score = coatt + sizee ## השלימו את הקוד
  if result_score == 1 or result_score == 2:  ## השלימו את הקוד
     return "fast"
  elif result_score == 3 or result_score == 4:          ## השלימו את הקוד
    return "medium"
  else:
    return "slow"


## כתבו קוד שיבדוק את הפונקציות שכתבתם מעלה
result = perd_model("short", "small")
print(result)
````

עכשיו אנו צריכים לעדכן את שמות הפונקציות לחישוב ניקוד הפרמטרים שלנו במקום המתאים בקוד הממשק הבא, החלק הרלוונטי מסומן בהערה - שימו לב, לאחר שנסיים לעדכן , נרציץ ונקבל את המודל שלנו עובד.

````python
# @title ממשק המודל
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Interface using IPywidgets
param1_dropdown = widgets.Dropdown(
    options=list(shelter_df[param1_name].unique()),
    description=param1_name+":"
)

param2_dropdown = widgets.Dropdown(
    options=list(shelter_df[param2_name].unique()),
    description= param2_name+":"
)

output_text = widgets.Output()
output_image = widgets.Image(value=open('default_image.jpg', 'rb').read(), format='jpg',layout=widgets.Layout(width='150px'))

def calculate_and_display_result(change):
    with output_text:
        clear_output(wait=True)

        selected_param1 = param1_dropdown.value
        selected_param2 = param2_dropdown.value

        #השלימו כאן את הקריאה לפונקציה של חישוב ניקוד הפרמטר הרלוונטי
        # השלימו במקום XXX את שם הפונקציה שלכם לחישוב ניקוד עבור הפרמטר הראשון והשני
        score_param1 = coatL(selected_param1)
        score_param2 = sizeL(selected_param2)

        result_category = perd_model(score_param1, score_param2)
        print(f"Dog adoption time is {result_category}")

        # Display images based on the result
        image_path = f'{result_category}_adoption.jpg'
        output_image.value = open(image_path, 'rb').read()

# Link dropdowns to the function
param1_dropdown.observe(calculate_and_display_result, names='value')
param2_dropdown.observe(calculate_and_display_result, names='value')


# Display the interface
display(widgets.VBox([param1_dropdown, param2_dropdown, output_text, output_image]))

# Initial display
calculate_and_display_result(None)
````
