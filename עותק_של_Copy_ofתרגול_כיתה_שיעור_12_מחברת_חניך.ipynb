**השמה לתוך df: אז איך עושים את זה?**
</font>
<br>
על מנת לשנות ערך מסויים, ניצטרך לגשת אליו ישירות ולשנות אותו, למשל כך:<br>

artistsdf.loc[0,'popularity'] = 100
<br>
ומה שעשינו זה למעשה לקחת את האומן בשורה 0 בdf ולשנות את הפופולריות שלו ל100.

בואו נראה את זה בקוד:

</div>

```python
print("line 0 before change:\n")
print(artistsdf.loc[0])
print()
artistsdf.loc[0, 'popularity'] = 100

print("line 0 after change:\n")
print(artistsdf.loc[0])
```

<font color='#7BAE7F'>

**השמה לתוך df:**
</font><br>

</div>

```python
# השלימו קטע קוד
print("line 0 before change:\n")
print(artistsdf.loc[0])
print()
artistsdf.loc[0, 'popularity'] = 0

print("line 0 after change:\n")
print(artistsdf.loc[0])
```
