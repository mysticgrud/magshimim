# בסרטון ראיתם שמקובל לחלק את אינדקס מסת הגוף (BMI) לקטגוריות הבאות:

<div dir="rtl">

# ממשו פונקציה שמחזירה את הweight_category לפי ה-BMI.
# השתמשו בתמונה מעל כדי לדעת את השמות והטווחים.

# קראו לפונקציה שלכם weight_category

```python
#השלימו כאן קטע קוד
#------------
def weight_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    elif 30 <= bmi < 34.9:
        return "Obese"
    else:
        return "Extreamly Obese"
#------------
```

<div dir="rtl">

# בדקו מה הקטגוריה של בן אדם עם נתונים של גובה: 167 ס"מ ומשקל 68 ק"ג. השתמשו בפונקציות שכתבתם

```python
#השלימו את הקוד כאן
height = 167
weight = 68

category = weight_category(calculate_bmi(height, weight))

print(category)
```

<div dir="rtl">

# ממשו את הפונקציות שיצרתם קודם באמצעות LAMBDA והציגו איך הן עובדות על אותה הדוגמה הנתונה של גובה ומשקל (גובה 167 ס"מ ומשקל 68 ק"ג)

```python
#השלימו את הקוד כאן

calculate_bmi_02 = lambda height, weight: weight / ((height/100) ** 2)

weight_category_02 = lambda bmi: "Underweight" if bmi < 18.5 else "Normal weight" if 18.5 <= bmi < 24.9 else "Overweight" if 25 <= bmi < 29.9 else "Obese" if 30 <= bmi < 34.9 else "Extremely Obese"

category = weight_category_02(calculate_bmi_02(height, weight))

print (category)
```

<div dir="rtl">

# הוסיפו לדאטה של העיירה עמודה בשם weight_category לפי הפונקציות שכתבתם קודם

# אתם יכולים לבחור להוסיף את המידע באמצעות הפונקציות ה"ארוכות" שכתבתם או לפי הלמדה.
# הציגו את העמודה בצורה ויזואלית לבחירתכם, אם זה גרף עמודות שימו לב לסדר העמודות

```python
#השלימו כאן קטע קוד להוספת העמודה-
filtered_df['weight_category'] = filtered_df.apply(lambda row: weight_category_02(calculate_bmi_02(row['height'], row['weight'])), axis=1)
filtered_df
```

# ישנה שיטה להפוך משתנים קטוגרים מסויימים למשתנים כמותיים. ראינו זאת במודל אפגר. כך אנו יכולים לעשות מניפולציות שלא מתאפשרות עם משתנה קטגורי בד"כ.

# באו ננסה לעשות זאת גם כאן.

# נראה כי יש לנו משתנים שמכילים אמת ושקר , שזה משתנה קטגוריאלי, אבל הוא יכול להיות גם בינארי!
# ניתן לקבוע כי אמת יסומן בספרה 1, ושקר יסומן בספרה 0.


# זוכרים איך לשנות ערכים בDF?
# בואו ננסה לעשות זאת עבור המשתנים 	has_crohn,is_smoker. המירו את כל ערכי ה-True בעמודות ל-1 ואת כל ערכי ה-False בעמודות ל-0. כך תהפכו את העמודות מעמודות בהן  יש משתנים איכותיים לעמודות בהן יש משתנים כמותיים
