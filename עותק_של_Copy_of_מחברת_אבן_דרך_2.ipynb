לאחר שביצעתם חקירה קצרה על הנתונים, בוודאי שמתם
לב שעבור
המשתנים הקטגוריים כמו "האם האדם מעשן" (smoke) מופיעות המילים yes, no. <BR> על מנת שנוכל לעבוד על הנתונים, נרצה שהנתונים שלנו יתוארו ע"י מספרים ולא ע"י מילים. לכן עליכם להמיר כל yes למספר 1, ולהמיר כל no למספר 0.
רמז: היעזרו בפונקציה map.
<br>
חשבו מה המטרה של הפיכת משתנה קטגורי לכמותי (זוכרים את מודל אפגר?)

````python
df.head()
````

````python
# השלימו את הקוד כאן
#df['smoke'].replace({'yes': 1, 'no': 0}, inplace=True) זה קוד שיותר אהבתי אבל בקבוצה אמרו שחובה להשתמש בmap
def new_thing(info):
    if info == 'yes':
        return 1
    elif info == 'no':
        return 0

df['smoke'] = list(map(new_thing, df['smoke']))
df['alco'] = list(map(new_thing, df['alco']))
df['active'] = list(map(new_thing, df['active']))
df['cardio'] = list(map(new_thing, df['cardio']))
````

<div dir=rtl align=right>
<font  size="5">
<div dir="rtl"><font color='#6400E4'><b> 3.5 מציאת מדדי מרכז (שאלת אתגר)</b></font></div>
</div>

<div dir="rtl">
<font  size="4">
על מנת להבין טוב יותר את הנתונים והשפעתם על הסיכוי לחלות במחלת לב, מצאו את מדדי המרכז של המשתנים הכמותיים עבור אנשים שנמצאים בסיכון לחלות במחלת לב ועבור אנשים שאינם נמצאים בסיכון.
<br>
האם ניתן לראות הבדל בין אלו שנמצאים בסכנה לחלות ועבור אלו שלא?
<br>
שימו לב, ייתכן ותיתקלו בשגיאות, אל תפחדו! חיקרו על השגיאות באינטרנט, והיעזרו במדריך במידת הצורך.
</div>

````python
# השלימו את הקוד כאן
danger = df['cardio'] == 1
no_danger = df['cardio'] == 0

danger_df = df[danger]
nodanger_df = df[no_danger]

print("danger weight stuff: ", danger_df['weight'].mean(), danger_df['weight'].mode(), danger_df['weight'].median())
print("danger height stuff: ", danger_df['height'].mean(), danger_df['height'].mode(), danger_df['height'].median())
print("danger age stuff: ", danger_df['age'].mean(), danger_df['age'].mode(), danger_df['age'].median())
print("danger ap_hi stuff: ", danger_df['ap_hi'].mean(), danger_df['ap_hi'].mode(), danger_df['ap_hi'].median())
print("danger ap_lo stuff: ", danger_df['ap_lo'].mean(), danger_df['ap_lo'].mode(), danger_df['ap_lo'].median())
print("danger cholesterol stuff: ", danger_df['cholesterol'].mean(), danger_df['cholesterol'].mode(), danger_df['cholesterol'].median())
print("danger gluc stuff: ", danger_df['gluc'].mean(), danger_df['gluc'].mode(), danger_df['gluc'].median())

print("no danger weight stuff: ", nodanger_df['weight'].mean(), nodanger_df['weight'].mode(), nodanger_df['weight'].median())
print("no danger height stuff: ", nodanger_df['height'].mean(), nodanger_df['height'].mode(), nodanger_df['height'].median())
print("no danger age stuff: ", nodanger_df['age'].mean(), nodanger_df['age'].mode(), nodanger_df['age'].median())
print("no danger ap_hi stuff: ", nodanger_df['ap_hi'].mean(), nodanger_df['ap_hi'].mode(), nodanger_df['ap_hi'].median())
print("no danger ap_lo stuff: ", nodanger_df['ap_lo'].mean(), nodanger_df['ap_lo'].mode(), nodanger_df['ap_lo'].median())
print("no danger cholesterol stuff: ", nodanger_df['cholesterol'].mean(), nodanger_df['cholesterol'].mode(), nodanger_df['cholesterol'].median())
print("no danger gluc stuff: ", nodanger_df['gluc'].mean(), nodanger_df['gluc'].mode(), nodanger_df['gluc'].median())
````

<br>
על סמך החקירה שביצעתם בשלב 2 ביחרו שישה משתנים (שאחד מהם יהיה 'cardio') אשר לפי דעתכם יש להם השפעה על הסיכוי לחלות במחלת לב.

````python
# השלימו את הקוד כאן
df=df.dropna()
df2 = df[['gender','age', 'cardio', 'weight', 'ap_hi', 'ap_lo']]
df2.dropna()
````

````python
#@title הריצו את התא על מנת להפעיל את הסיווג
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Check data validity
try:
    df2
except NameError:
    df2 = None

if df2 is None:
    print("Please make sure that you created a dataframe with the name: df2 that includes 6 columns")
elif 'cardio' not in df2.columns:
    print('Please make sure that one of the columns of df2 is "cardio"')
elif len(df2.columns) != 6:
    print("Please make sure that you have chosen 6 columns for df2")
elif df.isnull().values.any():
    print("Machine Learning models don't know how to handle NaN values, please make sure that your table does not have any NaN values")
elif len(df) < 67_000:
    print("You have removed too many rows from the dataframe, the limit is 3000 rows that can be removed")
else:

    # Prepare features and labels
    all_features = df2.drop('cardio', axis=1)
    true_labels = df2['cardio']

    # Split the data into training and testing sets
    train_test = train_test_split(all_features, true_labels, test_size=0.15, random_state=1)
    features_train, features_test, true_train_labels, true_test_labels = train_test

    # Define and train the KNN model
    knn_model = KNeighborsClassifier(n_neighbors=7)
    knn_model.fit(features_train, true_train_labels)

    # Predict labels for the test set
    pred_labels = knn_model.predict(features_test)

    # Generate and display confusion matrix
    cf_matrix = confusion_matrix(true_test_labels, pred_labels)
    ax = sns.heatmap(cf_matrix/sum(cf_matrix), annot=True, fmt='.2%', cmap='Blues')
    ax.invert_yaxis()
    ax.invert_xaxis()
    ax.set(xlabel='Predicted', ylabel='True')

    # Calculate model accuracy
    accuracy = sum(true_test_labels == pred_labels) / len(true_test_labels) * 100
    print(f"Your model accuracy is: {accuracy:.3f}%")
````
