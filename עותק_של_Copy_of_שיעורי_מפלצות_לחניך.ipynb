# איזה מפלצות תחמניות , לא פשוט למצוא בינהם הבדלים בכלל וככה קשה יותר גם למצוא איך להילחם בהם.

# בואו ננסה כיוון חדש, ננסה להבין את הגודל שלהם. אנו מציעים לקבוע 3 סוגי גדלים (קטן, בינוני וגדול).

# על סמך הנתונים בחרו משתנה שלפיו תקבעו את הקטגוריה מתאימה (את הגודל המתאים, בחרו את המשתנה הכי מתאים לדעתכם) והסתכלו על הנתונים וחשבו/קבעו מה טווח שמתאים לכל קטגוריה של גודל לפי המשתנה שבחרתם. המלצה לא להתעלם מסוג המפלצת גם.

# הוסיפו עמודה חדשה עם הגודל הרלוונטי לכל מפלצת והציגו את הכמויות של כל קטגורייה. מספרית וגרפית. (מי שלא זוכר איך להוסיף עמודות חזרו לשיעור 13 או חפשו באיטנרנט)

````python
# הוספת עמודה בהתאם לקטגוריה
def category(num):
    if num < 0.34:
        return 'small'
    elif num < 0.51:
        return 'mid'
    else:
        return 'big'

df['size'] = df['bone_length'].apply(category)
````

````python
# כמות המפלצות לפי הגודל שלהן
x = df['size'].value_counts()
````

````python
# גרף עמודות שמייצג את הכמות לפי גודל
sns.set(style="whitegrid")
sns.countplot(x="size", data=df, palette="Set3")
````

````python
#@title באיזה קטגוריה יצא לכם שיש הכי הרבה מפלצות?
variable_name = "בינוני" #@param ["\u05E7\u05D8\u05DF", "\u05D1\u05D9\u05E0\u05D5\u05E0\u05D9", "\u05D2\u05D3\u05D5\u05DC"]
````

# יופי לפחות התקדמנו קצת , ויש לנו תמונה יותר כללית על גדלי המפלצות איתם אנחנו מתמודדים.

# באו נראה מה לגבי הנשמה שלהם, הוסיפו עמודה חדשה שתציג אמת ושקר האם יש או אין נשמה למפלצת. נקבע כי כל המפלצות שיש להם ערך מעל 0.5 הן בעלי נשמה ,כל מה שמתחת אין.

# השתמשו בלולאה ותנאי

# הוסיפו תצוגה ויזואלית שמציגה כמויות לפי סוגי המפלצת וחלוקצת צבעים - אם יש או אין נשמה.השתמשו ב [מאפיין](https://seaborn.pydata.org/generated/seaborn.displot.html) -  multiple="stack"


````python
# השלימו קטע קוד
def has_soul(num):
  if (num>0.5):
    return True
  else:
    return False
df['soul_word'] = df['has_soul'].apply(has_soul)
df
````

````python
# השלימו את הגרף
sns.displot(data = df, x = 'type', hue = 'soul_word', multiple = 'stack')
````

# באו נלמד עוד דרך מקוצרת יותר לביצוע פעולה זו , איך לעשות זאת בשורה אחת.

```
df['has_soul_bool'] = df['has_soul'].apply(lambda x: True if x > 0.5 else False)
```

# נשתמש בפעולה APLLY שמוסיפה את מה שנבקש עבור כל שורה לעמודה המבוקשת.
# בתוך ההוספה נייצר מיני פונקציה שכבר פגשתם בשם lambda
#X  הוא הפרמטר הנוכחי בשורה הספציפית. נקבע שהוא אמת ורק אם הוא גדול מ0.5 ערכו יהיה שווה לשקר.

[ראו כאן דוגמה נוספת עם הסברים](https://www.geeksforgeeks.org/python-pandas-apply/)

# לאחר שראיתם את הדוגמה וההסבר, צרו באותו אופן עמודה חדשה שמכילה קטגוריית שיער - שיער קצר ושיער ארוך. כל מה שקטן מ0.5 יחשב כשיער קצר וכל מה שגדול מ0.5 יחשב כשיער ארוך.

````python
#כתבו כאן את הקוד
df['hair_length_bool'] = df['hair_length'].apply(lambda x: True if x > 0.5 else False)
````
